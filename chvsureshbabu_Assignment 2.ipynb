```python
#Q1) Write a python Class Circle which has the following variables and functions.

#Variables:
#pi_value = 3.14 (private variable)
#radius 

#functions:
#area_calculator
#circumference_calculator


#Create an object of Class circle and calculate its area and circumference.
```

```python
#define class
class Circle:
    def __init__(self, radius):
        self.__pi_value = 3.14  # private variable
        self.radius = radius
    
    def area_calculator(self):
        return self.__pi_value * self.radius ** 2
    
    def circumference_calculator(self):
        return 2 * self.__pi_value * self.radius

# Create an object of Class Circle
circle = Circle(5)

# Calculate and print the area and circumference
area = circle.area_calculator()
circumference = circle.circumference_calculator()

print(f"Area: {area}")
print(f"Circumference: {circumference:.2f}")
```

```python
#Q4) 	Read the Housing.csv file in your notebook and perform the following operations.
#i)	Calculate average and median price per square foot area (assume area column’s unit is sq ft )
#ii)	Calculate median price per bedroom. 
#iii)	Calculate average price per sq ft for furnished, semi-furnished and unfurnished homes.
#iv)	Are homes with guestrooms AND main road location costlier than remaining houses? Analyze the data using pandas to come up with your answer.
#v)	 Convert the values in air conditioning columns to 0 and 1 using lambda function
#vi)	Create a new column with the name price_index. Assign it 1 if the price of house is higher than average price else assign it value 0
#vii)	Sort the data by area, bathrooms and bedrooms column in descending order
```

```python
import pandas as pd
#Read the Housing.csv file in your notebook .
data=pd.read_csv("Housing.csv")
data.head()
#i)	Calculate average and median price per square foot area (assume area column’s unit is sq ft )
#data.shape
#data.index
#data.columns
#data[['price','area']].head()
#data[(data['price']>1000000)& (data ['bedrooms']>2)].head(10) #where with and 
#data[(data['furnishingstatus'] == 'furnished') | (data['mainroad'] == 'yes')].head(10) #where with or
data['dt_per_sqft']=data['price']/data['area']
data_median=data['dt_per_sqft'].median()
print(f"The median of per sqt is :",data_median)
data_mean=data['dt_per_sqft'].mean()
print(f"The Average of per sqt is :",data_mean)


#ii)	Calculate median price per bedroom. 

data_median_per_bedroom=data.groupby(by=['bedrooms'])['price'].median()
print("median price per bedroom :",data_median_per_bedroom)


#iii)	Calculate average price per sq ft for furnished, semi-furnished and unfurnished homes.
dt_furnishingstatus_mean=data.groupby(by=['furnishingstatus'])['price'].mean()
print(dt_furnishingstatus_mean)

#iv)Are homes with guestrooms AND main road location costlier than remaining houses? Analyze the data using pandas to come up with your answer.

# Filter homes with guestrooms and main road location
guestroom_mainroad_homes = data[(data['guestroom'] == 'yes') & (data['mainroad'] == 'yes')]
# Calculate the average price of homes with guestrooms and main road location
average_price_guestroom_mainroad = guestroom_mainroad_homes['price'].mean()
print(f"average price for homes with guest room & on Main road:{average_price_guestroom_mainroad:.2f}")
# Calculate the average price of remaining homes
remaining_homes = data[~((data['guestroom'] == 'yes') & (data['mainroad'] == 'yes'))]
average_price_remaining_homes = remaining_homes['price'].mean()
print(f"average price for homes without  guest room & not on Main road:{average_price_remaining_homes:.2f}")

if average_price_guestroom_mainroad > average_price_remaining_homes:
    print("Homes with guestrooms and main road location are costlier than remaining houses.")
else:
    print("Homes with guestrooms and main road location are not costlier than remaining houses.")


#v)	 Convert the values in air conditioning columns to 0 and 1 using lambda function
data['airconditioning'] = data['airconditioning'].apply(lambda x: 1 if x == 'yes' else 0)
data.head()


#vi)Create a new column with the name price_index. Assign it 1 if the price of house is higher than average price else assign it value 0

# Calculate the average price
average_price = data['price'].mean()
# Create price_index column
data['price_index'] = data['price'].apply(lambda x: 1 if x > average_price else 0)

print(data.head())
#vii)Sort the data by area, bathrooms and bedrooms column in descending order

data.sort_values(by=['area','bathrooms','bedrooms'],ascending=[False,False,False],inplace=True)
data.head()
```
